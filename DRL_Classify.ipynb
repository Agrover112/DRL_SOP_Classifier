```python
def vis_classification(X_test,y_test,y_pred):
    #Doing PCA to reduce the dimensions
    from sklearn.decomposition import PCA
    pca=PCA(n_components=2)
    X_embdedded=pca.fit_transform(X_test)
    #Colours based on Predicted words
    import seaborn as sns
    sns.scatterplot(X_embedded[0],X_embedded[1],hue=y_pred,s=170,marker='.',alpha=1)
    sns.scatterplot(X_embedded[0],X_embedded[1],hue=y_test,s=270,marker='o',alpha=0.4)    
    plt.title("Result of classification")
    plt.legend(loc='best')
    for i in range(len(y_pred)):
            if y_test[i]!=y_pred[i]:
                    plt.text(X_embedded.values[i][0],X_embedded.values[i][1]+2.0,s="Mistake")
    
        

vis_classification(X_test,y_test,y_pred)
```

```python
def plt_resize_text(labelsize, titlesize):
    ax = plt.subplot()
    for ticklabel in (ax.get_xticklabels()):
        ticklabel.set_fontsize(labelsize)
    for ticklabel in (ax.get_yticklabels()):
        ticklabel.set_fontsize(labelsize)
    ax.xaxis.get_label().set_fontsize(labelsize)
    ax.yaxis.get_label().set_fontsize(labelsize)
    ax.title.set_fontsize(titlesize)

plt.title("Total Dataset",fontsize=24)
data=Wt[(Wt["Frequency"]<12)&(Wt["Frequency"]>3)].sort_values("Frequency",ascending=False)
sns.barplot(x="Words",y="Frequency",data=data)
plt.rcParams['figure.figsize']=30,20
plt_resize_text(16,40)
plt.xticks(np.arange(len(data)), data.Words[:len(data)], rotation=60)
plt.xlabel("Words",fontsize=30)
plt.ylabel("Frequency",fontsize=30)
```
